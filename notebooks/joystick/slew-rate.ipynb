```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

```python
from wpimath.filter import SlewRateLimiter


y_filter = SlewRateLimiter(0.01, 0.0)
```

```python
x_filter = SlewRateLimiter(5*1e4, 0.0)
x_filter.reset(0)
t = np.arange(-1.0,1.0, 0.02)
x_in = np.heaviside(t, 0.5)

filtered = []

for x in np.nditer(x_in):
    filtered.append(x_filter.calculate(x))

plt.plot(t, x_in)
plt.plot(t, filtered)
```
