```python
%matplotlib notebook
from matplotlib import pyplot as plt
from wpimath.geometry import Translation2d
```

```python
a = Translation2d(1, 1)
b = Translation2d(1, 3)

print("plus", a+b)
print("distance", b.distance(a))
```

```python
x = [1, 1]
y = [1, 3]
a = ["A", "B"]
fig, ax = plt.subplots()
ax.set_xlim(0, 2)
ax.set_ylim(0, 4)
ax.scatter(x, y)
for i, txt in enumerate(a):
    ax.text(x[i]+0.1, y[i], txt)
```
